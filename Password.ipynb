```python
password_tuple=np.array(data)
```

```python
password_tuple
```

```python
import random
random.shuffle(password_tuple)
```

```python
x=[label[0] for label in password_tuple] 
y=[label[1] for label in password_tuple] 
```

```python
x
```

```python
def word_divide_char(inputs):
    character=[]
    for i in inputs:
        character.append(i)
    return character
```

```python
word_divide_char('lamborghin1')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer=TfidfVectorizer(tokenizer=word_divide_char)
```

```python
X=vectorizer.fit_transform(x)
print(vectorizer.vocabulary_)
```

```python
X.shape
```

```python
vectorizer.get_feature_names()
```

```python
first_document_vector=X[0]
first_document_vector
```

```python
first_document_vector.T.todense()
```

```python
df=pd.DataFrame(first_document_vector.T.todense(),index=vectorizer.get_feature_names(),columns=['TF-IDF'])
df.sort_values(by=['TF-IDF'],ascending=False)
```

Logistic regreesion is used as it converts linear relation into a Sigmoid one. Which solves the problem of outliers(-ve probability and greater than 1 probabbility)


```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape
```

```python
from sklearn.linear_model import LogisticRegression 
```

```python
clf=LogisticRegression(random_state=0,multi_class='multinomial')
```

```python
clf.fit(X_train,y_train)
```

```python
dt=np.array(['%123abcd2365'])
```

```python
pred=vectorizer.transform(dt)
```

```python
clf.predict(pred)
```

```python
y_pred= clf.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score 
```

```python
cm=confusion_matrix(y_test,y_pred)
print(cm)
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
